```python
import glob
import os
import re

path = os.path.abspath("/group/ag_abi/seiler/seqan3_explicit_links/seqan3/include/")

valid = re.compile(r'^\s*(\*|//!|/\*!|\*/)[\s]*(\x08rief|\\details|[^\\]).*')
replace = re.compile("(?<!\\\\link )(seqan3::[^\s\.]+)")
```

```python
for filename in glob.iglob(os.path.join(path, "**", "*.hpp"), recursive=True):
    filename = "/group/ag_abi/seiler/seqan3_explicit_links/seqan3/include/seqan3/alignment/aligned_sequence/aligned_sequence_concept.hpp"
    print(filename)
    buffer = []
    with open(filename, 'r') as f:
        buffer = f.readlines()
#     with open(filename, 'w') as f:
    header = False
    for line in buffer:
        if "\\file" in line:
            header = True
        elif header and "*/" in line:
            header = False
        elif not header and valid.search(line):
            print("BEFORE\n" + line)
            line = re.sub(replace, "\\\\link \g<0> \\\\endlink", line)
            print("AFTER\n" + line)
#         f.write(line)
    break
```
