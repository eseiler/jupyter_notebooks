```python
import os
import pandas as pd
import re

path = os.path.abspath("/group/ag_abi/seiler/IBF/textSize/")
```

```python
row_list = []
for b in [64,256,1024,8192]:
    for k in [12,13,14,15,16,17,18,19,20]:
        filename = os.path.join(path, "{}_{}.counts".format(b, k))
        row_dict = {}
        row_dict["b"] = b
        row_dict["k"] = k
        with open(filename, "r") as f:
            counts = []
            total = 0
            for line in f:
                [binNo, count] = line.split("\t")
                if binNo.strip() == "Overall":
                    total = int(count.strip())
                else:
                    counts.append(int(count.strip()))
            if len(counts) != b:
                raise Exception('Bins are missing')
            row_dict["ratio"] = "{0:,.2f}".format(sum(counts) / total)
            row_dict["total"] = "{0:,}".format(total)
        row_list.append(row_dict)
                
```

```python
df = pd.DataFrame(row_list)
```

```python
df = df.sort_values(by=['b', 'k'])
```

```python
df = df[["b",
         "k", 
         "ratio",
         "total"]]
```
