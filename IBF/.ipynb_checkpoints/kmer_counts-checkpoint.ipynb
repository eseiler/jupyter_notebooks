```python
import os
import pandas as pd
import re

path = os.path.abspath("/group/ag_abi/seiler/caroline_SRA/fasta/")
```

```python
row_list = []
filename = os.path.join(path, "counts.txt")
with open(filename, "r") as f:
    while True:
        row_dict = {}
        file = f.readline();
        count = f.readline();
        if not count: break
        [original, split] = re.findall('\d+', file)
        row_dict["ID"] = int(original.strip())
        row_dict["bin"] = int(split.strip())
        row_dict["count"] = int(count.strip())
        row_list.append(row_dict)
```

```python
df = pd.DataFrame(row_list)
```

```python
df = df.sort_values(by=['ID', 'bin'])
```

```python
df = df[["ID",
         "bin", 
         "count"]]
```
