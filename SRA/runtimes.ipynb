```python
import os
import pandas as pd

path = os.path.abspath("/group/ag_abi/seiler/caroline_SRA/SRA_Search/timings")
```

```python
row_list = []
for k in [19, 20]:
    for w in [24, 25, 26, 40, 41, 42, 43, 44]:
        filename = os.path.join(path, "k" + str(k) + "_" + "w" + str(w) + "_dna5.tsv")
        with open(filename, "r") as file:
            row_dict = {}
            row_dict["k"] = k
            row_dict["w"] = w
            for line in file:
                if line.startswith("Overall sequences"):
                    row_dict["Reads"] = line.split(':')[1].strip()
                elif line.startswith("Average HashTime per sequence:"):
                    row_dict["HashTime in ns per read"] = line.split(':')[1].strip()
                elif line.startswith("Average ThresholdTime per sequence:"):
                    row_dict["ThresholdTime in ns per read"] = line.split(':')[1].strip()
                else:
                    continue
            row_dict["Factor"] = "{0:.2f}".format(float(row_dict["ThresholdTime in ns per read"]) / float(row_dict["HashTime in ns per read"]))
            row_list.append(row_dict)
```

```python
df = pd.DataFrame(row_list)
```

```python
df = df[["k",
         "w", 
         "HashTime in ns per read", 
         "ThresholdTime in ns per read", 
         "Factor", 
         "Reads"]]
```
