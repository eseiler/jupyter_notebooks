```python
import os
import pandas as pd
import re

path = os.path.abspath("/group/ag_abi/seiler/caroline_SRA/kmer_counts/")
```

```python
row_list = []
for k in [19,20]:
    for w in [k, 23, 24, 25, 26, 27, 40, 41, 42, 43, 44, 45]:  
        filename = os.path.join(path, "{}_{}_dna5.tsv".format(k, w))
        with open(filename, "r") as f:
            for line in f:
                row_dict = {}
                [binNo, count] = line.strip().split('\t')
                if binNo != "Overall":
                    binNo = int(binNo)
                row_dict["k"] = k
                row_dict["w"] = w
                row_dict["bin"] = binNo
                row_dict["count"] = int(count)
                row_list.append(row_dict)
```

```python
df = pd.DataFrame(row_list)
```

```python
df = df[["k",
         "w", 
         "bin",
         "count"]]
```

```python
df.to_csv(os.path.join(path, "complete_counts_dna5.tsv"), sep='\t', index=False)
```
