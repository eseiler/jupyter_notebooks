```python
import os
import pandas as pd

path = os.path.abspath("/group/ag_abi/seiler/caroline_SRA/kmer_counts/")
bins = 50
```

```python
bin = 0
filename = os.path.join(path, str(bin) + ".log")
with open(filename, "r") as file:
    for line in file:
        line = line.lstrip()
        if line.startswith("No. of k-mers below min. threshold"):
            print(line.split(':')[1].strip())
```

```python
row_list = []
for b in range(bins):
    filename = os.path.join(path, str(b) + ".log")
    with open(filename, "r") as file:
        row_dict = {}
        row_dict["Bin"] = str(b)
        for line in file:
            line = line.lstrip()
            if line.startswith("No. of k-mers below min. threshold"):
                row_dict["No. of k-mers below min. threshold"] = line.split(':')[1].strip()
            elif line.startswith("No. of k-mers above max. threshold"):
                row_dict["No. of k-mers above max. threshold"] = line.split(':')[1].strip()
            elif line.startswith("No. of unique k-mers"):
                row_dict["No. of unique k-mers"] = line.split(':')[1].strip()
            elif line.startswith("No. of unique counted k-mers"):
                row_dict["No. of unique counted k-mers"] = line.split(':')[1].strip()
            elif line.startswith("Total no. of k-mers"):
                row_dict["Total no. of k-mers"] = line.split(':')[1].strip()
            elif line.startswith("Total no. of reads"):
                row_dict["Total no. of reads"] = line.split(':')[1].strip()
            elif line.startswith("Total no. of super-k-mers"):
                row_dict["Total no. of super-k-mers"] = line.split(':')[1].strip()
            else:
                continue
        row_list.append(row_dict)
```

```python
df = pd.DataFrame(row_list)
```

```python
df = df[["Bin",
         "No. of k-mers below min. threshold", 
         "No. of k-mers above max. threshold", 
         "No. of unique k-mers", 
         "No. of unique counted k-mers", 
         "Total no. of k-mers", 
         "Total no. of reads", 
         "Total no. of super-k-mers"]]
```
